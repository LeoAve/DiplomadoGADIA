```python
#Conectarse a la fuente de datos (entre comillas va el copiar ruta del archivo salary data)
salarios = pd.read_csv ("/content/sample_data/Salary_Data (1).csv")
salarios.head(10)
```

```python
#Cambiar nombre de columnas
salarios= salarios.rename(columns={"Salary":"Salario"})

salarios= salarios.rename(columns={"YearsExperience":"AñosDeExperiencia"})
```

##Regresión líneal

```python
#Gráfica de dispersión. Salario vs. experiencia
#La primera parte es del eje X y la segunda del eje Y
Y = salarios["Salario"]
X = salarios["AñosDeExperiencia"]
#Scatter es un tipo de gráfica de dispersión, el primer valor es el eje X y el segundo es el eje Y para poder generar la gráfica.

plt.scatter(X,Y,color = "red")

#Titulo del grafico
plt.title("Salarios vs AñosDeExperiencia")

#Titulo eje X
plt.xlabel("Años De Experiencia")

#Titulo eje Y
plt.ylabel("Salarios")
```
