**FOR**

Bucle for El bucle for se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, conjunto o cadena de caracteres). Se ejecuta una vez por cada elemento en la secuencia.

```python
#Imprimir todos los números de una secuencia de datos

secuencia= [1,2,3,4,5]

for i in secuencia :
  print(i)
```

```python
#Clasificar los numeros que son pares e impares de la lista secuencia

for i in secuencia:
    if i %2 ==0:
      print(f"el numero {i} es par")
    else:
       print(f"el numero {i} es impar")
```

```python
# Programa para aplicar un descuento del 10% a una lista de precios

def aplicar_descuento(precios):
    return [p * 0.9 for p in precios]

if __name__ == "__main__":
    precios = [100, 150, 200, 250, 300]
    precios_con_descuento = aplicar_descuento(precios)
    print("Precios originales:", precios)
    print("Precios con descuento:", precios_con_descuento)
```

```python

precio = [100,150,200,250,300]
precio_con_descuento = []


for i in precio:
    nuevo_precio = i*0.9
    precio_con_descuento.append
    print(f"Precio full es: {i}, pero con descuento es: {nuevo_precio}")
```

```python
import random

numero_secreto = random.randint(1, 10)

while True:

    numero = input("Adivina el # secreto (1-10): ")

    if not numero.isdigit():
        print("No es valido. Debes ingresar # entre 1-10.")
        continue

    numero = int(numero)

    if numero < 1 or numero > 10:
        print("# No es valido, Debes ingresar # entre 1-10.")
        continue

    if numero == numero_secreto:
        print("El # es correcto.")
        break
    else:
        print("Fallaste, prueba con otro #5")
```
